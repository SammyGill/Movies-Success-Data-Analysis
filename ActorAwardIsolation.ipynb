```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import patsy
import statsmodels.api as sm
import scipy.stats as stats
```

```python
df = pd.read_csv('database.csv')
```

```python
#col_actor_awards = [len(dict_films)]
actor_awards = df.loc[((df['Award'] == "Actor") | (df['Award'] == "Actress")), 'Film']
```

```python
arr_film_one = []
for i in range(0,actor_awards.shape[0]):
    try:
        if actor_awards[i] not in arr_film_one:
            arr_film_one.append(actor_awards[i])
    except KeyError:
        pass
    
total_movies = []
for i in range(0,df.shape[0]):
    try:
        if df.Film[i] not in total_movies:
            total_movies.append(df.Film[i])
    except KeyError:
        pass
    
arr_film_zero = list(arr_film_one)
for i in range(0,len(total_movies)):
    if total_movies[i] not in arr_film_zero:
        arr_film_zero.append(total_movies[i])
```

```python
arr_awards_one = [0] * len(arr_film_one)
for i in range(0, len(arr_film_one)):
    arr_awards_one[i]+=actor_awards[actor_awards == arr_film_one[i]].shape[0]

arr_awards_zero = list(arr_awards_one)
for i in range(0, len(total_movies)-len(arr_film_one)):
    arr_awards_zero.append(0)
```

```python
col_awards_one = pd.DataFrame({'actor_awards': arr_awards_one})
col_film_one = pd.DataFrame({'film': arr_film_one})
df_actor_awards_one = pd.concat([col_awards_one, col_film_one],axis = 1)
df_actor_awards_one
```

```python
col_awards_zero = pd.DataFrame({'actor_awards': arr_awards_zero})
col_film_zero = pd.DataFrame({'film': arr_film_zero})
df_actor_awards_zero = pd.concat([col_awards_zero, col_film_zero],axis = 1)
df_actor_awards_zero
```
